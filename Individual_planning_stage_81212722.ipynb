```r
library(tidyverse)
library(tidymodels)
library(ggplot2)
```

```r
players <- read_csv("players.csv")
players
```

```r
glimpse(players)
head(players)
```

```r
players |>
  summarise(across(everything(), ~ sum(is.na(.)))) |>
  pivot_longer(everything(),
               names_to = "variable",
               values_to = "missing_count")

players <- players |>
  mutate(
    subscribe  = as_factor(subscribe),
    experience = as_factor(experience),
    gender     = as_factor(gender)
  )

players |> summary()

players
```

```r
players |>
  summarise(
    mean_age = mean(Age, na.rm = TRUE),
    sd_age = sd(Age, na.rm = TRUE),
    mean_played = mean(played_hours, na.rm = TRUE),
    sd_played = sd(played_hours, na.rm = TRUE)
  )
players
```

```r
players |>
  count(experience, sort = TRUE)
players |>
  count(gender, sort = TRUE)

players |>
  count(subscribe) |>
  mutate(percent = round(n / sum(n) * 100, 1))
```

```r
players |>
  group_by(experience) |>
  summarise(mean_played = mean(played_hours, na.rm = TRUE))
```

```r
Played_Hour_diff <- ggplot(players, aes(x = played_hours)) +
  geom_histogram(binwidth = 5, fill = "blue", color = "white") +
  labs(title = "Played Hours Differences", 
       x = "Played Hours", 
       y = "Count"
      )

Subscribed_time <- ggplot(players, aes(x = subscribe, y = played_hours, fill = subscribe)) +
  geom_boxplot() +
  labs(title = "Playtime difference by Subscription Status", 
       x = "Subscribed", 
       y = "Played Hours"
      )

Proportion <- ggplot(players, aes(x = experience, fill = subscribe)) +
  geom_bar(position = "fill") +
  labs(title = "Proportion of Subscription by Experience", 
       y = "Proportion"
      )

Age <- ggplot(players, aes(x = subscribe, y = Age, fill = subscribe)) +
  geom_boxplot() +
  labs(title = "Age by Subscription", 
       x = "Subscribed", 
       y = "Age"
      )
Played_Hour_diff
Subscribed_time
Proportion
Age
```

```r
set.seed(2025)
players_split <- initial_split(players, prop = 0.8, strata = subscribe)
players_train <- training(players_split)
players_test  <- testing(players_split)

```

```r
players_recipe <- recipe(subscribe ~ experience + played_hours + Age, data = players_train)
players_recipe
```

```r
knn_spec <- nearest_neighbor(neighbors = 5) |>
  set_engine("kknn") |>
  set_mode("classification")
```

```r
log_workflow <- workflow() |>
  add_model(log_spec) |>
  add_recipe(players_recipe)
```

```r
players_folds <- vfold_cv(players_train, v = 5, strata = subscribe)
```

```r
log_res <- fit_resamples(
  log_workflow,
  resamples = players_folds,
  metrics = metric_set(accuracy, spec)
)
```

```r
collect_metrics(log_res)
```

```r
set.seed(2025)
final_fit <- fit(log_workflow, data = players_train)

final_results <- predict(final_fit, players_test) |>
  bind_cols(players_test)

final_metrics <- final_results |>
  metrics(truth = subscribe, estimate = .pred_class)

final_metrics
```
